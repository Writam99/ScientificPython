```
plt.plot(x,y, color = 'r', linestyle = '-', marker = 'o', ms = 16, lw = 4
         , label = 'Temperature')
plt.legend(loc = 2)
```

loc = location

```
plt.plot(x,y, color = 'r', linestyle = '-', marker = 'o', ms = 16, lw = 4, label = 'Temp.')
plt.legend(loc = 2)
plt.xlabel('X-AXIS', size = 20)
plt.ylabel('Y-AXIS', size = 20)
plt.title("This is my first graph", size = 20)
plt.suptitle("FIG. 1", size = 22)
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,10,20,30])
plt.tick_params(axis = 'x', labelcolor = 'g', labelsize = 16)
plt.xlim(0,5)
plt.ylim(0,40)
```
